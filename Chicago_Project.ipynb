```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# === LOAD DATA ===
df_schools = pd.read_csv("Chicago_Public_Schools_-_School_Profile_Information_SY2324_20250409.csv")
df_realestate = pd.read_csv("RealestateDIM.csv")
df_crimes = pd.read_csv("crimes_chicago_2023_to_present_with_zipcodes.csv").dropna()
df_popinc = pd.read_csv("PopulationANDIncome.csv")

# Clean population column
df_popinc['Population'] = (
    df_popinc['Population'].str.replace(',', '', regex=False)
    .astype(float)
    .fillna(lambda x: x.mean())
)
```

```python
# === CONFIG ===
COLUMNS_INCOME = ['Income_<25k', 'Income_>25K_<50K', 'Income_>50K_<100K', 'Income_>100K']

# Crime categories mapping
CRIME_CATEGORIES = {
    'ViolentCrimes': [
        'ROBBERY', 'BATTERY', 'HOMICIDE', 'CRIMINAL SEXUAL ASSAULT', 'CRIM SEXUAL ASSAULT',
        'ASSAULT', 'KIDNAPPING', 'ARSON', 'SEX OFFENSE', 'WEAPONS VIOLATION', 'STALKING',
        'INTIMIDATION', 'HUMAN TRAFFICKING', 'OFFENSE INVOLVING CHILDREN'
    ],
    'PropertyCrimes': [
        'CRIMINAL DAMAGE', 'THEFT', 'BURGLARY', 'MOTOR VEHICLE THEFT', 'CRIMINAL TRESPASS',
        'DECEPTIVE PRACTICE'
    ],
    'DrugCrimes': ['NARCOTICS', 'OTHER NARCOTIC VIOLATION']
}
# Everything else will be 'Other'
```

```python
# === MEDIAN ENROLLMENT & GRADUATION ===
def median_by_zip(df, zip_col, target_col, new_name):
    med_df = df.groupby(zip_col)[target_col].median().reset_index()
    med_df.columns = ['ZipCode', new_name]
    med_df[new_name] = med_df[new_name].fillna(round(med_df[new_name].mean(), 1))
    return med_df

med_enrollment = median_by_zip(df_schools, 'Zip', 'College_Enrollment_Rate_School', 'MedEnrollment')
med_graduation = median_by_zip(df_schools, 'Zip', 'Graduation_Rate_School', 'MedGraduation')
```

```python
# === CLEAN REALESTATE DATA ===
df_realestate.rename(columns={'ZIP Code(s)': 'ZipCode'}, inplace=True)
df_realestate = (
    df_realestate
    .assign(ZipCode=df_realestate['ZipCode'].str.split(','))
    .explode('ZipCode')
    .reset_index(drop=True)
)

# Convert income columns to numeric
for col in COLUMNS_INCOME:
    df_realestate[col] = (
        df_realestate[col]
        .apply(lambda x: 0.06 if isinstance(x, str) and '%' in x else x)
        .astype(str)
        .str.replace(',', '', regex=False)
        .astype(float)
    )

# Average by zip
avg_by_zip = df_realestate.groupby('ZipCode')[
    ['Forclosure2023'] + COLUMNS_INCOME
].mean().reset_index()

# Add population and foreclosure rate
df_pop = df_popinc[['ZipCode', 'Population']].copy()

# Ensure ZipCode is int for both before merging
avg_by_zip['ZipCode'] = avg_by_zip['ZipCode'].astype(int)
df_pop['ZipCode'] = df_pop['ZipCode'].astype(int)

df_realestate_merged = avg_by_zip.merge(df_pop, on='ZipCode', how='left')
df_realestate_merged['Foreclosure_rate'] = (
    df_realestate_merged['Forclosure2023'] / df_realestate_merged['Population'] * 10000
)
```

```python
# === CRIME CLASSIFICATION ===
# Create a reverse lookup dictionary for faster mapping
crime_map = {crime: category for category, crimes in CRIME_CATEGORIES.items() for crime in crimes}

df_crimes.rename(columns={'zip_code': 'ZipCode'}, inplace=True)
df_crimes['CrimeCategory'] = df_crimes['Primary Type'].map(crime_map).fillna('Other')
df_crime_counts = df_crimes.groupby(['ZipCode', 'CrimeCategory']).size().unstack(fill_value=0).reset_index()
```

```python
# === MERGE ALL DATA ===
df_full = (
    df_pop[['ZipCode']]
    .merge(med_enrollment, on='ZipCode', how='outer')
    .merge(med_graduation, on='ZipCode', how='outer')
    .merge(df_realestate_merged, on='ZipCode', how='outer')
    .merge(df_crime_counts, on='ZipCode', how='outer')
)
```

```python
# === CALCULATE INCOME SCORE ===

df_full['IncomeScore'] = (
    df_full['Income_<25k'] * 1 +
    df_full['Income_>25K_<50K'] * 3 +
    df_full['Income_>50K_<100K'] * 5 +
    df_full['Income_>100K'] * 6
)
```

```python
# === FILL MISSING VALUES ===
columns_to_fill = [
    'MedEnrollment', 'MedGraduation', 'Foreclosure_rate',
    'IncomeScore', 'DrugCrimes', 'Other', 'PropertyCrimes', 'ViolentCrimes'
]

for col in columns_to_fill:
    # Ensure the column is numeric
    df_full[col] = pd.to_numeric(df_full[col], errors='coerce')
    # Fill missing values with the median
    df_full[col] = df_full[col].fillna(df_full[col].median())
```

```python
# === NORMALIZE SELECTED COLUMNS ===
scaler = MinMaxScaler()
norm_cols = ['MedEnrollment', 'MedGraduation', 'Forclosure2023', 'IncomeScore',
             'DrugCrimes', 'Other', 'PropertyCrimes', 'ViolentCrimes']
df_full[norm_cols] = scaler.fit_transform(df_full[norm_cols])
```

```python
# === CALCULATE NEIGHBORHOOD SCORE ===
weights = {
    'MedEnrollment': 0.2,
    'MedGraduation': 0.2,
    'Foreclosure_rate': -0.2,
    'IncomeScore': 0.2,
    'DrugCrimes': -0.1,
    'Other': -0.03,
    'PropertyCrimes': -0.12,
    'ViolentCrimes': -0.35
}
df_full['NeighborhoodScore'] = sum(df_full[col] * w for col, w in weights.items())
df_full['Rank'] = df_full['NeighborhoodScore'].rank(ascending=False, method='first')
```

```python
# === SAVE RESULTS ===
df_full.to_csv('Chicago_NeighborhoodScores.csv', index=False)
df_full.head()
```
